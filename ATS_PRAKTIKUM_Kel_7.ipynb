```python
def hitung_efisiensi(input_energi, output_energi):
    efisiensi = output_energi / input_energi
    return efisiensi

# Cara memanggil fungsi
hasil = hitung_efisiensi(500, 400)
print("Efisiensi turbin:", hasil)
```

```python
# Fungsi untuk menghitung efisiensi termal
def hitung_efisiensi_termal():
    # Meminta input dari pengguna
    energi_masuk = float(input("Masukkan energi masuk (Joule): "))
    energi_keluar = float(input("Masukkan energi keluar (Joule): "))

    # Menghitung efisiensi dengan rumus (energi keluar / energi masuk) Ã— 100
    efisiensi = (energi_keluar / energi_masuk) * 100

    # Menampilkan hasil efisiensi
    print("Efisiensi termal turbin gas adalah:", efisiensi, "%")

# Memanggil fungsi
hitung_efisiensi_termal()
```

```python
# Simulasi pengisian tangki bahan bakar
kapasitas = 0  # mulai dari 0 liter

while kapasitas <= 100:
    print(f"Tangki berisi {kapasitas} liter.")

    if kapasitas == 80:
        print("WARNING: Hampir penuh!")
    elif kapasitas == 100:
        print("TANGKI PENUH - Pengisian berhenti.")
        break  # hentikan loop saat tangki penuh

    kapasitas += 10  # naikkan 10 liter setiap iterasi
```

```python
# Fungsi untuk menghitung total pemakaian listrik
def hitung_total(pemakaian_list):
    return sum(pemakaian_list)

# Simulasi pengisian data energi
pemakaian = []
total = 0
jam = 0

while total <= 900:
    jam += 1
    energi = int(input(f"Masukkan pemakaian listrik jam ke-{jam} (Wh): "))
    pemakaian.append(energi)
    total = hitung_total(pemakaian)

    if total >= 800 and total < 900:
        print(" WARNING: Pemakaian listrik hampir mencapai batas harian!")
    elif total >= 900:
        print(" Batas pemakaian tercapai! Harap matikan beberapa peralatan.")
        break
    else:
        print(f"Total sementara: {total} Wh\n")

print(f"Total pemakaian hari ini: {total} Wh")
```
