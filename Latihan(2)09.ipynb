```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Aiman_df = pd.read_csv('/content/customers.csv')
```

```python
Aiman_df.describe()
```

```python
import matplotlib.pyplot as plt

Aiman_df['age'].hist()
plt.title("Data Customer")
plt.xlabel("Tahun Customer")
plt.ylabel("Jumlah Customer")
plt.show()
```

```python
# Ringkasan statistik
Aiman_df.describe()

#Rata-rata umur
mean_age = Aiman_df['age'].mean()
print("Rata-rata umur:", mean_age)

# Modus umur
mode_age = Aiman_df['age'].mode()
print("Modus umur:", mode_age)
```

```python
import matplotlib.pyplot as plt

# Umur
Aiman_df['age'].hist()
plt.title("Umur customer")
plt.show()

# Zip
Aiman_df['zip_code'].hist()
plt.title("Zip")
plt.show()
```

```python
# Korelasi
corr_hw = Aiman_df[['age', 'zip_code']].corr()
print(corr_hw)

# Kovarian
cov_hw = Aiman_df[['age', 'zip_code' ]].cov()
print(cov_hw)
```
